```python
import spacy
import pandas as pd
from gensim.models import Phrases
from collections import Counter
nlp = spacy.load('en')
```

```python
doc = open('bitcoin.txt', encoding="utf8").read()
```

```python
doc = doc.replace('\n',' ')
```

```python
btcWP = nlp(doc)
```

```python
btcWP_list = []
for sentence in btcWP.sents:
    print(sentence)
    btcWP_list.append(str(sentence))
```

use Gensim for phrase modeling

```python
lemmas = [token.lemma_ for token in btcWP if not token.is_stop]
lemmas
```

```python
sentence_stream = [doc.split(" ") for sentence in btcWP.sents]
bigram = Phrases(sentence_stream, min_count=5, threshold=10)
print(bigram[sentence_stream[0]])
df = pd.DataFrame(bigram[sentence_stream[0]])
```

```python
counts = Counter()
for token in btcWP:
    counts[token.orth_] += 1 # Equivalently, token.text
counts.most_common
```
